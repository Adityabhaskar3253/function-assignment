 **1.What is the difference between a function and a method in Python?**

 -> In Python, functions and methods are both callable objects, but they differ in terms of where and how they are defined and used:

**1. Function**

A function is a block of code that is defined using the def keyword and is designed to perform a specific task. It is not bound to any object.

    Definition: A function is defined independently of any object or class.
    Usage: A function is called directly by its name, passing the required arguments.
    Scope: It can be called from anywhere in the program as long as it's in scope.


```python
#Example of a function:
def greet(name):
    return f"Hello, {name}!"

# Calling the function
print(greet("Alice"))
```

**2. Method**

A method is a function that is associated with an object or a class. It is a function that operates on an instance of a class (an object) or the class itself (for class methods).

    Definition: A method is defined within a class and is bound to the instances (or class) of that class.
    Usage: A method is called on an instance or class using dot notation.
    Self: For instance methods, the first argument is typically self, which refers to the instance of the object that the method is being called on. For class methods, it is usually cls (referring to the class itself).

```python
#Example of a method:
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

# Creating an instance of the class
greeter = Greeter()

# Calling the method on the instance
print(greeter.greet("Alice"))
```

**6.Explain the concept of generators in Python and how they are defined?**

-> **How Generators Wor**k

    When a generator function is called, it does not execute immediately. Instead, it returns a generator object.
    The yield statement pauses the function and saves its state, allowing it to resume from where it left off when next() is called on the generator object.
    When the generator is exhausted, it raises a StopIteration exception.

  **Defining Generators**

Generators are defined using functions with one or more yield statements.

**Example:**

    def my_generator():
    yield 1
    yield 2
    yield 3

    gen = my_generator()
    print(next(gen))  # Output: 1
    print(next(gen))  # Output: 2
    print(next(gen))  # Output: 3
  # next(gen)  # Raises StopIteration


**Key Features**

    Lazy Evaluation: Generators produce items on demand, saving memory.
    State Preservation: Generators maintain their state between yield calls.
    Single Use: Once exhausted, a generator cannot be reused.

Key benefits of using generators:

**1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.**

->
 Here is a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list:
def sum_of_evens(numbers):
Returns the sum of all even numbers in the given list.

   :param numbers: List of integers
   
   :return: Sum of even integers  
    
 return sum(num for num in numbers if num % 2 == 0)

```python
def sum_of_even_numbers(numbers):
    """
    Returns the sum of all even numbers in the given list.

    :param numbers: List of integers
    :return: Sum of even integers
    """
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Example usage
numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print(result)
```

**4.Write a Python function that checks if a given number is prime or not from 1 to 200.**

 -> Here is a Python function that checks if a given number from 1 to 200 is prime or not and prints the prime numbers within that range

```python
def is_prime(n):
    """
    Checks if the given number is prime.

    :param n: The number to check
    :return: True if the number is prime, False otherwise
    """
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to the square root of n
        if n % i == 0:
            return False
    return True

def print_primes_up_to_200():
    """
    Prints all prime numbers from 1 to 200.
    """
    for num in range(1, 201):
        if is_prime(num):
            print(num)

# Example usage
print_primes_up_to_200()
```

**5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**


-> Here is a Python iterator class that generates the Fibonacci sequence up to a specified number of terms:

```python
class FibonacciIterator:
    def __init__(self, terms):
        """
        Initializes the Fibonacci iterator with the specified number of terms.

        :param terms: The number of terms in the Fibonacci sequence to generate
        """
        self.terms = terms
        self.current_term = 0
        self.prev, self.curr = 0, 1  # Fibonacci starts with 0, 1

    def __iter__(self):
        """
        Returns the iterator object itself.
        """
        return self

    def __next__(self):
        """
        Returns the next term in the Fibonacci sequence.

        Raises StopIteration when the specified number of terms is reached.
        """
        if self.current_term >= self.terms:
            raise StopIteration

        # Return the current Fibonacci number
        fibonacci_number = self.prev
        # Update the values for the next Fibonacci number
        self.prev, self.curr = self.curr, self.prev + self.curr
        self.current_term += 1

        return fibonacci_number

# Example usage
fib = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers
for num in fib:
    print(num)
```

**6.Write a generator function in Python that yields the powers of 2 up to a given exponent.**


Here is a Python generator function that yields the powers of 2 up to a given exponent:

```python
def powers_of_two(exponent):
    """
    Yields the powers of 2 up to the given exponent.

    :param exponent: The maximum exponent for powers of 2
    """
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage
for power in powers_of_two(5):  # Powers of 2 from 2^0 to 2^5
    print(power)
```

**7.Implement a generator function that reads a file line by line and yields each line as a string.**

-> Here is an implementation of a generator function that reads a file line by line and yields each line as a string:

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Use .strip() to remove trailing newline characters



```

How it works:

    The open(file_path, 'r') statement opens the file in read mode.
    The for line in file loop reads the file line by line.
    The yield keyword allows the function to return a line one at a time without reading the entire file into memory.
    The .strip() method removes any leading/trailing whitespace or newline characters from each line before yielding it.

11. an accounting routine used in a book shop. It works on a list with sub lists,
which look like this:
Order Number ,  Book Title and Author ,    Quantity  ,     Price per Item,

    34587 Learning Python, Mark Lutz 4 40.95
    98762 Programming Python, Mark Lutz 5 56.80
    77226 Head First Python, Paul Barry 3 32.95
    88112 Einfhrung in Python3, Bernd Klein 3 24.99

Write a Python program, which returns a list with 2-tuples. Each tuple consists of a
the order number and the product of the price per items and the quantity.

 The product
should be increased by 10,-  if the value of the order is smaller than 100,00 .

 Write
a Python program using lambda and map.

-> To solve this problem using lambda and map, we need to create a program that processes the list of orders, calculates the product of quantity and price per item, and applies a 10-euro increase if the order total is less than 100 euros

```python
# List of orders in the format (Order Number, Book Title and Author, Quantity, Price per Item)
orders = [
    (34587, "Learning Python, Mark Lutz", 4, 40.95),
    (98762, "Programming Python, Mark Lutz", 5, 56.80),
    (77226, "Head First Python, Paul Barry", 3, 32.95),
    (88112, "Einfhrung in Python3, Bernd Klein", 3, 24.99)
]

# Define a function to calculate the total price with the condition to add 10 if less than 100
def calculate_total(order):
    order_number, _, quantity, price_per_item = order
    total = quantity * price_per_item
    if total < 100:
        total += 10  # Add 10 if total is less than 100
    return (order_number, total)

# Use map and lambda to apply the function to the list of orders
order_totals = list(map(lambda order: calculate_total(order), orders))

# Print the result
for order in order_totals:
    print(order)
```
